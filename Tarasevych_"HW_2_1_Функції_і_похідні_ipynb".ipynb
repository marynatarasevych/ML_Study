Похілна функції 𝑓(𝑥)=10𝑥−5 = 10

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)  # Похідна f'(x) = 10

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10 * x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

plt.show()
```

При збільшенні годин навчання, оцінка буде вище. Залежність лінійна - значить що кожна година витрачена на навчання принесе вищу оцінку, немає точки, в якому цей приріст перестане працювати

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

# sin(x)
f_sin = sp.sin(x)
df_sin = sp.diff(f_sin, x)

# ln(x)
f_ln = sp.ln(x)
df_ln = sp.diff(f_ln, x)

# конвертуємо в numpy функції
f_sin_np = sp.lambdify(x, f_sin, 'numpy')
df_sin_np = sp.lambdify(x, df_sin, 'numpy')
f_ln_np = sp.lambdify(x, f_ln, 'numpy')
df_ln_np = sp.lambdify(x, df_ln, 'numpy')

xs = np.linspace(0.1, 10, 100)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(xs, f_sin_np(xs), color='blue', label='sin(x)')
ax1.plot(xs, df_sin_np(xs), color='green', label="cos(x)")
ax1.legend()
ax1.set_title('sin(x) та її похідна')

ax2.plot(xs, f_ln_np(xs), color='blue', label='ln(x)')
ax2.plot(xs, df_ln_np(xs), color='green', label="1/x")
ax2.legend()
ax2.set_title('ln(x) та її похідна')

plt.show()
```
