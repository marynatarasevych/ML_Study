Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
mean_vis = sum(visits)/len(visits)
mean_vis
```

```python
minus_mean_v = [(el - mean_vis)**2 for el in visits]
std_vis = np.sqrt(sum(minus_mean_v)/(len(visits)-1))
std_vis
```

```python
sum_vis = [ (sum_purchases[i] - mean_sum)*(visits[i] - mean_vis) for i in range(len(sum_purchases))]
cov_sv = sum(sum_vis)/(len(sum_purchases)-1)
cov_sv / (std_sum * std_vis)
```

```python
np.corrcoef(sum_purchases,visits)
```

Коефіцієнт кореляції Пірсона на наданих даних показує, що є слабка негативна кореляція між данними на рівні -0,29. Це значення ближче до 0, ніж до 1 - тому можемо сказати що дані майже не корелюють між собою.

```python
import matplotlib.pyplot as plt

plt.scatter(sum_purchases, visits)
plt.title('Scatter Plot of Purchases vs. Visits')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

Зробивши графік видно, що в сумі покупок є 2 викиди, які можуть викривляти кореляцію. Можливо, якщо їх прибрати, то кореляція стане більш вагомою.

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
threshold = mean_sum + 2 * std_sum
threshold
```

```python
sum_purchases = [ 316,  493,  367,  153,  224,  121,  322,  450,  196,  397,  281,
  239,  484,  124,  223,  448,  432,  255, 4310, 3524]
```

```python
keep_indices = [i for i, x in enumerate(sum_purchases) if x <= threshold]
keep_indices
```

```python
filtered_sum = [sum_purchases[i] for i in keep_indices]
filtered_vis = [visits[i] for i in keep_indices]
print (filtered_sum)
print (filtered_vis)
```

```python
np.corrcoef(filtered_sum,filtered_vis)
```

```python
plt.scatter(filtered_sum, filtered_vis)
plt.title('Scatter Plot of Purchases vs. Visits w/o outliers')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits');
```

Прибравши викиди, можна побачити на графіку, що немає залежності суми покупок від кількості візитів. Коефіцієнт кореляції став -0,37 (був -0,29). Хоча він і підвищився, проте все одно набагато менше 1, тому можна вважати, що кореляції немає.
